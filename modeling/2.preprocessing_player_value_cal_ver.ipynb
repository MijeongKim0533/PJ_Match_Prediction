```python
import pandas as pd
```

```python
# 파일 가져오기
players_df = pd.read_csv('./basic_data/players.csv')
players_df
```

```python
# 파일 가져오기
player_valuations_df = pd.read_csv('./basic_data/player_valuations.csv')
player_valuations_df
```

```python
# player_valuations_df 정렬
player_valuations_df = player_valuations_df.sort_values(['player_id', 'date'])
```

```python
# merge
merged_df = pd.merge(players_df, player_valuations_df, on='player_id')
```

```python
merged_df
```

```python
# 필요한 열만 선택
players_value_df = merged_df[['player_id', 'name', 'date', 'market_value_in_eur_y', 'current_club_id_y', 'player_club_domestic_competition_id']]
```

```python
players_value_df
```

```python
# 다시 정렬
players_value_df = players_value_df.sort_values(['player_id', 'date'])
```

```python
players_value_df
```

```python
# lead_date 생성
players_value_df['lead_date'] = players_value_df.groupby('player_id')['date'].shift(-1)
```

```python
players_value_df
```

```python
# 선수별로 lead_date 가장 최근날짜에 null값 생김 -> 현재 날짜로 대체
players_value_df = players_value_df.fillna('2024-05-03')
```

```python
players_value_df
```

```python
# null 확인
players_value_df.isnull().sum()
```

```python
# 컬럼 순서 변경
players_value_df = players_value_df[['player_id', 'name', 'date', 'lead_date', 'market_value_in_eur_y',
       'current_club_id_y', 'player_club_domestic_competition_id']]
```

```python
players_value_df
```

```python
# 날짜 형식으로 변경
players_value_df['date'] = pd.to_datetime(players_value_df['date'])
```

```python
players_value_df
```

```python
players_value_df.head()
```

```python
# 각 플레이어의 최소 날짜 구하기
min_dates = players_value_df.groupby('player_id')['date'].transform('min')
```

```python
# 최소 날짜에서 1000일 빼기
new_dates = pd.to_datetime(min_dates) - pd.DateOffset(days=1000)

# 변경된 날짜로 업데이트
players_value_df.loc[players_value_df.groupby('player_id')['date'].idxmin(), 'date'] = new_dates
```

```python
players_value_df.head()
```
