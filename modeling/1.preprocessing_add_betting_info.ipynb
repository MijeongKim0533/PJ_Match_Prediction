```python
new_order = ['Date', 'Team', 'Opponent', 'Team_goal', 'Oppo_goal', 'Referee', 'TY',
       'OY', 'TR', 'OR', 'B365TW', 'B365D', 'B365OW']
```

```python
result_df = pd.concat([df_1824, df_1824_op], axis=0)
```
